* 小明的分數跟各科系權重可以直接複製上面的文字使用
* 表格裡面的內容可以用for迴圈跑過個科系 取出該科系的權重 然後整理成一個二維的list

```python
import pandas as pd
```

```python
subject =  ["國文", "英文", "數甲", "數乙", "物理", "化學", "生物", "歷史", "地理", "公民"]
weights = {
    "土語系": [2, 2, 0, 1, 0, 0, 0, 1, 1, 0],
    "財政系": [1, 1.5, 0, 1.5, 0, 0, 0, 0, 0, 0],
    "金融系": [1, 1.5, 0, 2, 0, 0, 0, 0, 0, 0],
    "國貿系": [1.5, 2, 0, 2, 0, 0, 0, 0, 0, 0],
    "會計系": [1.5, 2, 0, 2, 0, 0, 0, 1, 1, 0],
    "法律系": [1, 2, 0, 1, 0, 0, 0, 1, 0, 1],
    "公行系": [1, 1.5, 0, 1.25, 0, 0, 0, 0, 0, 1.25],
    "應數系": [1, 1, 1.5, 0, 1, 1, 0, 0, 0, 0],
    "教育系": [1.5, 2, 0, 1.25, 0, 0, 0, 1, 1, 0]
}
scores = [86.5, 79.4, 88.2, 96.5, 86.3, 76.8, 96.1, 78.5, 82.4, 88.6]
```

```python
mydata = list()
for department in weights:
    mydata.append(weights[department])
mydata
```

```python
df = pd.DataFrame(mydata, columns=subject, index=weights.keys())
df
```

```python
df["加權分數"] = (df * scores).sum(1)
```
